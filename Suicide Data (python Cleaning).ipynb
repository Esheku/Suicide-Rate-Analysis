```python
#use while loop and if statements to clean generation column, because wrong generations were assigned to the years.

index = 0
while index < len(data):
    if data['year'][index] in list(range(1977, 1996)):
        data['generation'][index] = 'Millenials'
    else:
        data['generation'][index] = 'Generation Z'
    index += 1
data
```
